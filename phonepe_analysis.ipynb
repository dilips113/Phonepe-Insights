```python
import pandas as pd
import json
import os
```

```python
# --- TRANSACTION DATA ---

path = 'I:/Project/pulse/data/aggregated/transaction/country/india/state'
Agg_state_list = os.listdir(path)
Agg_state_list

trans_col = {'States':[], 'Years':[], 'Quarter':[], 'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for state in Agg_state_list:
    cur_states = path +"/"+ state +"/"
    agg_year_list = os.listdir(cur_states)
    #print(cur_states)

    for year in agg_year_list:
        cur_year = cur_states + year + "/"
        agg_file_list = os.listdir(cur_year)
        #print(cur_year)

        for file in agg_file_list:
            cur_file = cur_year+file
            #print(cur_file)
            data = open(cur_file,'r')
            D1 = json.load(data)

            for i in D1['data']['transactionData']:
                name = i['name']
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                trans_col["Transaction_type"].append(name)
                trans_col["Transaction_count"].append(count)
                trans_col["Transaction_amount"].append(amount)
                trans_col["States"].append(state)
                trans_col["Years"].append(year)
                trans_col["Quarter"].append(int(file.strip(".json")))

Agg_Trans = pd.DataFrame(trans_col)

Agg_Trans['States'] = Agg_Trans['States'].str.replace('-', ' ').str.replace('&', 'and')
Agg_Trans['States'] = Agg_Trans['States'].str.title() #To Captialise 


# --- INSURANCE DATA ---

path2 = 'I:/Project/pulse/data/aggregated/insurance/country/india/state'
agg_insur_list = os.listdir(path2)
agg_insur_list

insur_col = {'States':[],'Years':[],'Quarter':[],'Insurance_type':[],'Insurance_count':[],'Insurance_amount':[]}

for state in agg_insur_list:
    cur_states2 = path2+'/'+ state + '/'
    agg_year_list2 = os.listdir(cur_states2)

    for year in agg_year_list2:
        cur_year2 = cur_states2 + year +'/'
        agg_file_list2 = os.listdir(cur_year2)

        for file in agg_file_list2:
            cur_file2 = cur_year2 + file
            data2 = open(cur_file2,'r')
            D2 = json.load(data2)

            for i in D2['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i ['paymentInstruments'][0]['amount']
                insur_col["Insurance_type"].append(name)
                insur_col["Insurance_count"].append(count)
                insur_col["Insurance_amount"].append(amount)
                insur_col["States"].append(state)
                insur_col["Years"].append(year)
                insur_col["Quarter"].append(int(file.strip(".json")))

Agg_Insur = pd.DataFrame(insur_col)

Agg_Insur['States'] = Agg_Insur['States'].str.replace('-', ' ').str.replace('&', 'and')
Agg_Insur['States'] = Agg_Insur['States'].str.title() #To Captialise 


# --- USER DEVICE DATA ---

path3 = 'I:/Project/pulse/data/aggregated/user/country/india/state'
agg_user_list = os.listdir(path3)

user_col = {"States":[], "Years":[], "Quarter":[], "Brands":[],"Transaction_count":[], "Percentage":[]}

for state in agg_user_list:
    cur_states3 = path3+'/'+state+'/'
    agg_year_list3 = os.listdir(cur_states3)

    for year in agg_year_list3:
        cur_year3 = cur_states3+year+'/'
        agg_file_list3 = os.listdir(cur_year3)

        for file in agg_file_list3:
            cur_file3 = cur_year3+file
            data3=open(cur_file3,'r')
            D3 = json.load(data3)

            try:

                for i in D3['data']['usersByDevice']:
                    brand = i['brand']
                    count = i['count']
                    percentage = i['percentage']
                    user_col["Brands"].append(brand)
                    user_col["Transaction_count"].append(count)
                    user_col["Percentage"].append(percentage)
                    user_col["States"].append(state)
                    user_col["Years"].append(year)
                    user_col["Quarter"].append(int(file.strip(".json")))

            except:
                pass

Agg_User = pd.DataFrame(user_col)

Agg_User['States'] = Agg_User['States'].str.replace('-', ' ').str.replace('&', 'and')
Agg_User['States'] = Agg_User['States'].str.title() #To Captialise 

                
```

```python
#------- MAP INSURANCE -------#

path4 = 'I:/Project/pulse/data/map/insurance/hover/country/india/state/'
map_insur_list = os.listdir(path4)

map_insur_col = {'States': [], 'Years': [], 'Quarter': [], 'District': [], 'Insurance_count': [], 'Insurance_amount': []}

for state in map_insur_list:
    cur_states4 = path4 + "/" + state + "/"  
    map_year_list = os.listdir(cur_states4)

    for year in map_year_list:
        cur_year4 = cur_states4 + year + "/"
        map_file_list = os.listdir(cur_year4)

        for file in map_file_list:
            cur_file4 = cur_year4 + file
            data4=open(cur_file4,'r')
            D4 = json.load(data4)

            for i in D4['data']['hoverDataList']:
                    name = i['name']
                    count = i['metric'][0]['count']
                    amount = i['metric'][0]['amount']
                    map_insur_col["District"].append(name)
                    map_insur_col["Insurance_count"].append(count)
                    map_insur_col["Insurance_amount"].append(amount)
                    map_insur_col["States"].append(state)
                    map_insur_col["Years"].append(year)
                    map_insur_col["Quarter"].append(int(file.strip(".json")))

Map_Insur = pd.DataFrame(map_insur_col)

Map_Insur['States'] = Map_Insur['States'].str.replace('-', ' ').str.replace('&', 'and')
Map_Insur['States'] = Map_Insur['States'].str.title()



# ------- MAP TRANSACTION ------- #

path5 = 'I:/Project/pulse/data/map/transaction/hover/country/india/state/'
map_trans_list = os.listdir(path5)

map_trans_col = {'States': [], 'Years': [], 'Quarter': [], 'District': [], 'Transaction_count': [], 'Transaction_amount': []}

for state in map_trans_list:
    cur_states5 = path5 + state + "/"
    map_year_list = os.listdir(cur_states5)

    for year in map_year_list:
        cur_year5 = cur_states5 + year + "/"
        map_file_list = os.listdir(cur_year5)

        for file in map_file_list:
            cur_file5 = cur_year5 + file
            data5 = open(cur_file5, 'r')
            D5 = json.load(data5)

            for i in D5['data']['hoverDataList']:
                name = i['name']
                count = i['metric'][0]['count']
                amount = i['metric'][0]['amount']
                map_trans_col["District"].append(name)
                map_trans_col["Transaction_count"].append(count)
                map_trans_col["Transaction_amount"].append(amount)
                map_trans_col["States"].append(state)
                map_trans_col["Years"].append(year)
                map_trans_col["Quarter"].append(int(file.strip(".json")))

Map_Trans = pd.DataFrame(map_trans_col)

Map_Trans['States'] = Map_Trans['States'].str.replace('-', ' ').str.replace('&', 'and')
Map_Trans['States'] = Map_Trans['States'].str.title()



# ------- MAP USER ------- #

path6 = 'I:/Project/pulse/data/map/user/hover/country/india/state/'
map_user_list = os.listdir(path6)

map_user_col = {'States': [], 'Years': [], 'Quarter': [], 'District': [], 'RegisteredUsers': [], 'AppOpens': []}

for state in map_user_list:
    cur_states6 = path6 + state + "/"
    map_year_list = os.listdir(cur_states6)

    for year in map_year_list:
        cur_year6 = cur_states6 + year + "/"
        map_file_list = os.listdir(cur_year6)

        for file in map_file_list:
            cur_file6 = cur_year6 + file
            data6 = open(cur_file6, 'r')
            D6 = json.load(data6)

            for i in D6['data']['hoverData'].items():
                district = i[0]
                registereduser = i[1]["registeredUsers"]
                appopens = i[1]["appOpens"]
                map_user_col["District"].append(district)
                map_user_col["RegisteredUsers"].append(registereduser)
                map_user_col["AppOpens"].append(appopens)
                map_user_col["States"].append(state)
                map_user_col["Years"].append(year)
                map_user_col["Quarter"].append(int(file.strip(".json")))

Map_User = pd.DataFrame(map_user_col)

Map_User['States'] = Map_User['States'].str.replace('-', ' ').str.replace('&', 'and')
Map_User['States'] = Map_User['States'].str.title()
```

```python
#------- TOP INSURANCE -------#

path7 = 'I:/Project/pulse/data/top/insurance/country/india/state/'
top_insur_list = os.listdir(path7)

top_insur_col = {
    'States': [], 'Years': [], 'Quarter': [],
    'Entity_Level': [],
    'Entity_Name': [],
    'Insurance_count': [], 'Insurance_amount': []
}

for state in top_insur_list:
    cur_states7 = path7 + state + '/'
    year_list = os.listdir(cur_states7)

    for year in year_list:
        cur_year7 = cur_states7 + year + '/'
        file_list = os.listdir(cur_year7)

        for file in file_list:
            cur_file7 = cur_year7 + file
            with open(cur_file7, 'r') as data7:
                D7 = json.load(data7)

            # Process 'districts'
            for i in D7['data']['districts']:
                entity_level = "District"
                entity_name = i["entityName"]
                count = i["metric"]["count"]
                amount = i["metric"]["amount"]

                top_insur_col['Entity_Level'].append(entity_level)
                top_insur_col['Entity_Name'].append(entity_name)
                top_insur_col['Insurance_count'].append(count)
                top_insur_col['Insurance_amount'].append(amount)
                top_insur_col['States'].append(state)
                top_insur_col['Years'].append(year)
                top_insur_col['Quarter'].append(int(file.strip('.json')))

            # Process 'pincodes'
            for i in D7['data']['pincodes']:
                entity_level = "Pincode"
                entity_name = i["entityName"]
                count = i["metric"]["count"]
                amount = i["metric"]["amount"]

                top_insur_col['Entity_Level'].append(entity_level)
                top_insur_col['Entity_Name'].append(entity_name)
                top_insur_col['Insurance_count'].append(count)
                top_insur_col['Insurance_amount'].append(amount)
                top_insur_col['States'].append(state)
                top_insur_col['Years'].append(year)
                top_insur_col['Quarter'].append(int(file.strip('.json')))

Top_Insur = pd.DataFrame(top_insur_col)

Top_Insur['States'] = Top_Insur['States'].str.replace('-', ' ').str.replace('&', 'and')
Top_Insur['States'] = Top_Insur['States'].str.title()


#------- TOP TRANSACTION -------#

path8 = 'I:/Project/pulse/data/top/transaction/country/india/state/'
top_trans_list = os.listdir(path8)

top_trans_col = {
    'States': [], 'Years': [], 'Quarter': [],
    'Entity_Level': [],  
    'Entity_Name': [],
    'Transaction_count': [], 'Transaction_amount': []
}

for state in top_trans_list:
    cur_states8 = path8 + state + '/'
    year_list = os.listdir(cur_states8)

    for year in year_list:
        cur_year8 = cur_states8 + year + '/'
        file_list = os.listdir(cur_year8)

        for file in file_list:
            cur_file8 = cur_year8 + file
            with open(cur_file8, 'r') as data8:
                D8 = json.load(data8)

            for i in D8['data']['districts']:
                entity_level = "District"
                entity_name = i["entityName"]
                count = i["metric"]["count"]
                amount = i["metric"]["amount"]

                top_trans_col['Entity_Level'].append(entity_level)
                top_trans_col['Entity_Name'].append(entity_name)
                top_trans_col['Transaction_count'].append(count)
                top_trans_col['Transaction_amount'].append(amount)
                top_trans_col['States'].append(state)
                top_trans_col['Years'].append(year)
                top_trans_col['Quarter'].append(int(file.strip('.json')))

            for i in D8['data']['pincodes']:
                entity_level = "Pincode"
                entity_name = i["entityName"]
                count = i["metric"]["count"]
                amount = i["metric"]["amount"]

                top_trans_col['Entity_Level'].append(entity_level)
                top_trans_col['Entity_Name'].append(entity_name)
                top_trans_col['Transaction_count'].append(count)
                top_trans_col['Transaction_amount'].append(amount)
                top_trans_col['States'].append(state)
                top_trans_col['Years'].append(year)
                top_trans_col['Quarter'].append(int(file.strip('.json')))


Top_Trans = pd.DataFrame(top_trans_col)


Top_Trans['States'] = Top_Trans['States'].str.replace('-', ' ').str.replace('&', 'and')
Top_Trans['States'] = Top_Trans['States'].str.title()


#------- TOP USER -------#

path9 = 'I:/Project/pulse/data/top/user/country/india/state/'
top_user_list = os.listdir(path9)

top_user_col = {
    'States': [], 'Years': [], 'Quarter': [],
    'Entity_Level': [], 
    'Entity_Name': [],
    'Registered_Users': []
}

for state in top_user_list:
    cur_states9 = path9 + state + '/'
    year_list = os.listdir(cur_states9)

    for year in year_list:
        cur_year9 = cur_states9 + year + '/'
        file_list = os.listdir(cur_year9)

        for file in file_list:
            cur_file9 = cur_year9 + file
            with open(cur_file9, 'r') as data9:
                D9 = json.load(data9)

           
            for i in D9['data']['districts']:
                entity_level = "District"
                entity_name = i["name"]
                registered_users = i["registeredUsers"]

                top_user_col['Entity_Level'].append(entity_level)
                top_user_col['Entity_Name'].append(entity_name)
                top_user_col['Registered_Users'].append(registered_users)
                top_user_col['States'].append(state)
                top_user_col['Years'].append(year)
                top_user_col['Quarter'].append(int(file.strip('.json')))

            for i in D9['data']['pincodes']:
                entity_level = "Pincode"
                entity_name = i["name"]
                registered_users = i["registeredUsers"]

                top_user_col['Entity_Level'].append(entity_level)
                top_user_col['Entity_Name'].append(entity_name)
                top_user_col['Registered_Users'].append(registered_users)
                top_user_col['States'].append(state)
                top_user_col['Years'].append(year)
                top_user_col['Quarter'].append(int(file.strip('.json')))

Top_User = pd.DataFrame(top_user_col)

Top_User['States'] = Top_User['States'].str.replace('-', ' ').str.replace('&', 'and')
Top_User['States'] = Top_User['States'].str.title()
```
